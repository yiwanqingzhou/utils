```python
import os
path = "/home/bot/Downloads/rgb_training/1280x960"

fileList=os.listdir(path)
```

```python
fileList.sort(key=lambda s: int(s[s.find('_') + 1: s.find('.')]))
```

```python
for file in fileList:
    whole_path = path + os.sep + file
```

```python
for file in fileList:
    oldname = path + os.sep + file

    id = int(file[file.find('_') + 1: file.find('.')])
#     print(id)
    
    if id > 4:
        new_id = id -1
        new_file = file[0:file.find('_')+1]+str(new_id)+file[file.find('.'):]
#         print(new_file)
    
        newname = path + os.sep + new_file
        os.rename(oldname,newname)
        print(oldname,'======>',newname)
```

```python
d = {}

for file in fileList:
    oldname = path + os.sep + file
    
    tote_id = file[-6:-4]
    if tote_id[0]=='_':
        tote_id = tote_id[1:]
        
    if tote_id in d:
        d[tote_id] = d[tote_id] + 1
    else:
        d[tote_id] = 0
    
    new_file = tote_id + "_" + str(d[tote_id])+".pcd"
    newname = path + os.sep + new_file
    
    os.rename(oldname,newname)
    print(oldname,'======>',newname)
```
